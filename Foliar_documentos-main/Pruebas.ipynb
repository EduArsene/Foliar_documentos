Funcion para agregar codigo de capo

```python
def agregar_elemento(agregar, tag, tipo, texto=None):
    elemento = OxmlElement(tag)
    elemento.set(qn('w:fldCharType'), tipo)
    if texto:
        elemento.set(qn('xml:space'), 'preserve')
        elemento.text = texto
    agregar._r.append(elemento)
```

Manipulacion de encabezados

```python
#*************Para poder activar opcion de headres diferentes
def activar_pares_impares_diferentes(seccion):
    seccPr = seccion._sectPr
    if seccPr is None:
        seccPr = OxmlElement('w:sectPr')
        seccion._element.append(seccPr)
    
    # Eliminar la opción de "Primera página diferente" si existe
    for el in seccPr.findall(qn('w:titlePg')):
        seccPr.remove(el)
    
    # Activar la opción de "Páginas pares e impares diferentes"
    evenAndOddHeaders = OxmlElement('w:evenAndOddHeaders')
    seccPr.append(evenAndOddHeaders)
    
def configurar_idioma_parrafo(parrafo, lang_code):
    for run in parrafo.runs:
        rPr = run._element.get_or_add_rPr()
        lang = OxmlElement('w:lang')
        lang.set(qn('w:val'), lang_code)
        rPr.append(lang)
```

```python
def impar(parrafo):
    correr = parrafo.add_run()
    # Inicio primer bloque
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', 'PAGE') # PAGE
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '+ 1')  # + 1
    agregar_elemento(correr, 'w:fldChar', 'separate')      # separador
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '/2') # /2
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    #fin primer bloque
    agregar_elemento(correr, 'w:instrText', 'begin', ':')  # :
    #Inicio segundo bloque
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', 'PAGE') # PAGE
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '+ 1')  # + 1
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '/2') # /2
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '\*CardText \*Upper') # \*CardText \*Upper
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    #Fin segundo bloque
 
    
def foliarImpar(doc):
    seccion = doc.sections[0]
    header = seccion.header
    parrafo = header.paragraphs[0] if header.paragraphs else header.add_paragraph()
    parrafo.alignment = WD_PARAGRAPH_ALIGNMENT.RIGHT
    impar(parrafo)
    configurar_idioma_parrafo(parrafo, 'es-PE')
    
def foliarImparPie(doc):
    seccion = doc.sections[0]
    footer = seccion.footer
    parrafo = footer.paragraphs[0] if footer.paragraphs else footer.add_paragraph()
    parrafo.alignment = WD_PARAGRAPH_ALIGNMENT.RIGHT
    impar(parrafo)
    configurar_idioma_parrafo(parrafo, 'es-PE')
```

```python
#******* PARA PAR
def par(parrafo):
  
    correr = parrafo.add_run()
    # Inicio primer bloque
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', 'PAGE') # PAGE
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '')  # - 1
    agregar_elemento(correr, 'w:fldChar', 'separate')      # separador
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '*0.5') # /2
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    #fin primer bloque
    agregar_elemento(correr, 'w:instrText', 'begin', ':')  # :
    #Inicio segundo bloque
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', '=')  # =
    agregar_elemento(correr, 'w:fldChar', 'begin')         # {
    agregar_elemento(correr, 'w:instrText', 'begin', 'PAGE') # PAGE
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '')  # - 1
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '*0.5') # /2
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    agregar_elemento(correr, 'w:instrText', 'begin', '\*CardText \*Upper') # \*CardText \*Upper
    agregar_elemento(correr, 'w:fldChar', 'end')           # }
    #Fin segundo bloque
 
    
def foliarPar(doc):
    for seccion in doc.sections:
        seccion.different_first_page_header_footer = False
        seccion.odd_and_even_pages_header_footer = True   
    # Asegurar que haya al menos dos p�ginas
    doc.add_page_break()
    # Acceder al encabezado de las p�ginas pares de la primera secci�n
    seccion = doc.sections[0]
    cabecera = seccion.even_page_header
    parrafo = cabecera.paragraphs[0] if cabecera.paragraphs else cabecera.add_paragraph()
    parrafo.alignment = WD_PARAGRAPH_ALIGNMENT.RIGHT
    # Agregar campo de página en el encabezado
    par(parrafo)
    configurar_idioma_parrafo(parrafo, 'es-PE')
    
def foliarParPie(doc):
    for seccion in doc.sections:
        seccion.different_first_page_header_footer = False
        seccion.odd_and_even_pages_header_footer = True   
    # Asegurar que haya al menos dos p�ginas
    doc.add_page_break()
    # Acceder al encabezado de las p�ginas pares de la primera secci�n
    seccion = doc.sections[0]
    piePag = seccion.even_page_footer
    parrafo = piePag.paragraphs[0] if piePag.paragraphs else piePag.add_paragraph()
    parrafo.alignment = WD_PARAGRAPH_ALIGNMENT.RIGHT
    # Agregar campo de página en el encabezado
    par(parrafo)
    configurar_idioma_parrafo(parrafo, 'es-PE')
```

```python
import PyPDF2
def establecer_idioma_espanol(doc):
    for paragraph in doc.paragraphs:
        for run in paragraph.runs:
            run.font.language_id = 0x040A  # Español (Perú)
#funcion principal para crear
def crear_documento_foliado():
    #obtener informacion de la ventada principal para luego ponerla al costado
    root_x = root.winfo_x()
    root_y = root.winfo_y()
    root_width = root.winfo_width()
    numero_paginas = tk.IntVar()
    #funcion para poder selccionar el pdf principal
    def seleccionar_pdf():
        archivo_pdf = filedialog.askopenfilename(filetypes=[("Archivos PDF", "*.pdf")])
        if archivo_pdf:
            ruta_pdf.set(archivo_pdf)
            try:
                reader = PdfReader(archivo_pdf)
                numero_paginas.set(len(reader.pages))                
            except Exception as e:
                messagebox.showerror("Error", f"No se pudo leer el archivo PDF: {e}")
    #Funcion para seleccionar la ruta donde se va a guardar el archivo
    def seleccionar_ruta_guardar():
        extension = ".pdf" #if formato_var.get() == "PDF" else ".docx"
        archivo = filedialog.asksaveasfilename(defaultextension=extension,
                                               filetypes=[(f"Archivos {extension}", extension)])
        if archivo:
            ruta_guardar.set(archivo)
            aux_ruta_guardar.set(archivo)
            
    #Funcion para poder obtener la orientacion del archivo pdf
    def obtener_orientacion(ruta_pdf):
        with open(ruta_pdf, 'rb') as file:
            pdf_leer = PyPDF2.PdfReader(file)
            pagina = pdf_leer.pages[0]
            
            page_ancho = pagina.mediabox[2]
            page_altura = pagina.mediabox[3]
            
            if page_ancho > page_altura:
                return 'Horizontal'
            else:
                return 'Vertical' 
    def superponerPdf():
        archivo_abajo_datos = ruta_pdf.get()
        if not archivo_abajo_datos:
            messagebox.showerror("Error", "Seleccione un archivo PDF válido para superponer.")
            return

        try:
            pdfAbajoReader = PyPDF2.PdfReader(archivo_abajo_datos)

            #archivo_arriba = filedialog.askopenfilename(filetypes=[("Archivos PDF", "*.pdf")])
            archivo_arriba = aux_ruta_guardar.get() 
            #arc=archivo_arriba.replace(".pdf", "aux.pdf")
           # messagebox.showinfo("",archivo_arriba)
            if not archivo_arriba:
                return  # Se canceló la selección del segundo archivo

            with open(archivo_arriba, 'rb') as archivo_arriba:
                pdfArribaReader = PyPDF2.PdfReader(archivo_arriba)

                for pageNum in range(0, len(pdfArribaReader.pages)):
                    pdfAbajoReader.pages[pageNum].merge_page(pdfArribaReader.pages[pageNum])

                pdfWriter = PyPDF2.PdfWriter()

                for pageNum in range(0, len(pdfAbajoReader.pages)):
                    pageObj = pdfAbajoReader.pages[pageNum]
                    pdfWriter.add_page(pageObj)
                
                result_pdf_file = ruta_pdf.get()
                result_pdf_file = result_pdf_file.replace(".pdf", "-FOLIADO.pdf")
                if not result_pdf_file:
                    return  # Se canceló la selección de la ruta de guardado

                with open(result_pdf_file, 'wb') as output_pdf:
                    pdfWriter.write(output_pdf)

            messagebox.showinfo("Éxito", f"PDF superpuesto guardado en: '{result_pdf_file}'")

        except Exception as e:
            messagebox.showerror("Error", f"No se pudo superponer los PDF: {e}")   
                         
    def obtener_tamannio_paginas(pdf_path):
        try:
            alto_ = 0
            with open(pdf_path, 'rb') as archivo_pdf:
                lector_pdf = PyPDF2.PdfReader(archivo_pdf)
                numero_paginas = len(lector_pdf.pages)

                for num_pagina in range(numero_paginas):
                    pagina = lector_pdf.pages[num_pagina]
                    tamanio = pagina.mediabox
                    ancho_cm = (tamanio.upper_right[0] - tamanio.lower_left[0]) * 0.0352778
                    alto_cm = (tamanio.upper_right[1] - tamanio.lower_left[1]) * 0.0352778
                    alto_ = alto_cm
                    print(f"Tamaño de la página {num_pagina + 1}: {ancho_cm:.2f} x {alto_cm:.2f} cm")
                return alto_
        except Exception as e:
            print(f"No se pudo leer el archivo PDF: {e}")
       
    def crear_doc():
        #llamar a rutas, la principal y donde se guardará lo creado
        archivo_pdf = ruta_pdf.get()
        archivo_guardar = ruta_guardar.get()
        archivo_guardar.replace(".pdf","aux.pdf")
        
        if not archivo_pdf:
            messagebox.showerror("Error", "Seleccione un archivo PDF válido.")
            return

        if not archivo_guardar:
            messagebox.showerror("Error", "Seleccione una ruta para guardar el documento.")
            return

        try:
            numero_de_paginas = int(numero_paginas.get())
            if numero_de_paginas <= 0:
                raise ValueError("El número de páginas debe ser mayor a 0")
        except ValueError as e:
            messagebox.showerror("Error", str(e))
            return
        #variable para manipular el doc
        doc = Document()
        establecer_idioma_espanol(doc)
        for i in range(1, numero_de_paginas):
            doc.add_page_break()

        orientacion = obtener_orientacion(ruta_pdf.get())
        for section in doc.sections:
            section.header_distance = Cm(0.3) 
            if orientacion == "Horizontal":
                section.footer_distance = Cm(0.85)
                section.orientation = WD_ORIENT.LANDSCAPE
                section.page_width, section.page_height = section.page_height, section.page_width
            else:
                section.orientation = WD_ORIENT.PORTRAIT      
        try:            
            opcion_PE = opcion_pie_encabezado.get()        
            opcionIP = opcion_var.get()            
            opcion_AD = opcion_asc_desc.get()           
            if opcion_PE == "Encabezado":#caso Encabezado de pagina
               if opcionIP == "Impar" :
                   if opcion_AD == "Ascendente":foliarImpar(doc)
                   else: foliarImparDesc(doc,numero_de_paginas)
               elif opcionIP == "Normal":
                  if opcion_AD == "Ascendente":foliarNormal(doc)  
                  else: foliarNormalDesc(doc,numero_de_paginas)
                                             
            else:                        #caso PIE de pagina
                if opcionIP == "Impar":
                    if opcion_AD == "Ascendente":foliarImparPie(doc)
                    else: foliarImparPieDesc(doc,numero_de_paginas)
                elif opcionIP == "Normal": 
                    if opcion_AD == "Ascendente":foliarNormalPie(doc)  
                    else: foliarNormalPieDesc(doc,numero_de_paginas)                                                               

            if opcionIP != "Normal":
                for section in doc.sections:
                    activar_pares_impares_diferentes(section)
                                  
            archivo_docx_temp = "documento_temporal.docx"
            doc.save(archivo_docx_temp)
            ruta_guardar.set('')
            convert(archivo_docx_temp, archivo_guardar)
            os.remove(archivo_docx_temp)
                
            superponerPdf()
            os.remove(archivo_guardar)
        except ValueError as e:
            messagebox.showerror("Error", str(e))
             
            
##############
    style = ttk.Style()
    style.theme_use('clam')

    # Definir colores
    color_primario = "#2c3e50"
    #secondary_color = "#34495e"
    accent_color = "#3498db"
    texto_color = "#ecf0f1"
    hover_color = "#34495e"  

    style.configure('TFrame', background=color_primario)
    style.configure('TLabel', background=color_primario, foreground=texto_color, font=('Helvetica', 10))
    style.configure('TButton', background=accent_color, foreground=texto_color, font=('Helvetica', 10, 'bold'))
    style.map('TButton', background=[('active', '#2980b9')])
    
    # Configuración especial para Radiobutton
    style.configure('TRadiobutton', background=color_primario, foreground=texto_color, font=('Helvetica', 10))
    style.map('TRadiobutton', 
              background=[('active', hover_color)],
              foreground=[('active', texto_color)])

    style.configure('TEntry', foreground=color_primario, font=('Helvetica', 10))
    #crear ventana 
    ventana_config = tk.Toplevel(root)
    ventana_config.title("Crear Documento Foliado")
    # Establecer la posición de la nueva ventana a la derecha de la ventana principal
    ventana_width = 700  # Ancho de la nueva ventana
    ventana_height = 700  # Altura de la nueva ventana
    ventana_x = root_x + root_width  # Posición en X (a la derecha de la ventana principal)
    ventana_y = root_y  # Mantener la misma posición en Y
    # Aplicar la geometría calculada
    ventana_config.geometry(f"{ventana_width}x{ventana_height}+{ventana_x}+{ventana_y}")
    
    ventana_config.attributes("-topmost", True)
    ventana_config.configure(bg=color_primario)

    main_frame = ttk.Frame(ventana_config, padding="20 20 20 20")
    main_frame.pack(fill=tk.BOTH, expand=True)

    # Función para crear secciones
    #Muestra de foliacion para aydar a diferencia entre par e impar
    def mostrar_ayuda():
        ventana_imagen = tk.Toplevel()
        ventana_imagen.title("MUESTRA DE FOLIACION")
        ventana_imagen.attributes("-topmost", True)  
        #canvas = tk.Canvas(ventana_imagen, width=1179, height=678)
        alto = 1100
        ancho = 700
        canvas = tk.Canvas(ventana_imagen, width=alto, height=ancho)#("2424x1593")
        canvas.pack()
        imagen_original = Image.open("Muestra.png")
        
        imagen_redimensionada = imagen_original.resize((alto, ancho))  # Ajusta el tamaño según lo necesario
        imagen = ImageTk.PhotoImage(imagen_redimensionada)
        
        canvas.create_image(0, 0, anchor=tk.NW, image=imagen)
        canvas.image = imagen#largoxalto
        ventana_imagen.geometry(f"{alto}x{ancho}")
        
    
    def create_section(parent, titulo):
        section_frame = ttk.Frame(parent)
        section_frame.pack(fill=tk.X, pady=10)
        ttk.Label(section_frame, text=titulo, font=('Helvetica', 14, 'bold'),foreground="black").pack(anchor=tk.W)
        return section_frame
    # Sección de selección de archivo PDF
    pdf_section = create_section(main_frame, "Selección de Archivo PDF")
    ruta_pdf = tk.StringVar()
    

    
    tk.Entry(pdf_section, textvariable=ruta_pdf, state='readonly', width=50).pack(side=tk.LEFT, padx=(0, 10))
    imagen_bot = Image.open("btnExaminar.png")
    imagen_redimensionada = imagen_bot.resize((325, 120))  # Ajusta el tamaño según sea necesario
    imagen_boton = ImageTk.PhotoImage(imagen_redimensionada)
    # Crear el botón con la imagen
    btn_examinar = tk.Button(
        pdf_section, 
        bg=color_primario,
        border=0,
        image=imagen_boton,  # Asigna la imagen al botón
        command=seleccionar_pdf,
        width=280,
        height=85,
        activebackground=color_primario
    )
    btn_examinar.image = imagen_boton  # Necesario para evitar la recolección de basura de la imagen
    #btn_examinar.pack(pady=20)
    btn_examinar.pack(side=tk.LEFT, padx=10)
    
        
    pie_encabezado = create_section(main_frame, "Seleccionar tipo de borde")
    opcion_pie_encabezado = tk.StringVar(value="Encabezado")
    tk.Radiobutton(pie_encabezado, text="Encabezado", 
                   variable=opcion_pie_encabezado, 
                   value="Encabezado",
                   bg=color_primario,
                   font=("Helvetica", 10, "bold")).pack(side=tk.LEFT, padx=(0, 20))
    tk.Radiobutton(pie_encabezado, text="Pie de página",
                   variable=opcion_pie_encabezado, 
                   value="Pie de página",
                   bg=color_primario,
                   font=("Helvetica", 10, "bold")).pack(side=tk.LEFT)

    impares_pares = create_section(main_frame, "Foliación (páginas impares o normales)")
    opcion_var = tk.StringVar(value="Impar")
    tk.Radiobutton(impares_pares, text="Impar", variable=opcion_var, value="Impar").pack(side=tk.LEFT, padx=(0, 20))
    tk.Radiobutton(impares_pares, text="Normal", variable=opcion_var, value="Normal").pack(side=tk.LEFT)
    tk.Button(impares_pares, text="¿Cómo debe quedar la foliación?", command=mostrar_ayuda).pack(side=tk.RIGHT, padx=5)

    asc_desc = create_section(main_frame, "Foliación (ascendente o descendente)")
    opcion_asc_desc = tk.StringVar(value="Ascendente")
    tk.Radiobutton(asc_desc, text="Ascendente", variable=opcion_asc_desc, value="Ascendente").pack(side=tk.LEFT, padx=(0, 20))
    tk.Radiobutton(asc_desc, text="Descendente", variable=opcion_asc_desc, value="Descendente").pack(side=tk.LEFT)

    save_section = create_section(main_frame, "Guardar PDF Resultante")
    ruta_guardar = tk.StringVar()
    aux_ruta_guardar = tk.StringVar()  
    tk.Entry(save_section, textvariable=ruta_guardar, state='readonly', width=50).pack(side=tk.LEFT, padx=(0, 10))
    tk.Button(save_section, text="Guardar como", command=seleccionar_ruta_guardar).pack(side=tk.LEFT)

    boton_frame = tk.Frame(main_frame)
    boton_frame.pack(fill=tk.X, pady=20)

    tk.Button(boton_frame, text="Crear Documento", command=crear_doc).pack(side=tk.LEFT, padx=5)
    tk.Button(boton_frame, text="Cancelar", command=ventana_config.destroy).pack(side=tk.LEFT, padx=5)
    

def cambiar_formatos():
    def seleccionar_archivo():
        archivo = filedialog.askopenfilename(filetypes=[("Archivos de Word", "*.docx")])
        if archivo:
            ruta_archivo.set(archivo)

    def convertir_archivo():
        archivo_word = ruta_archivo.get()
        if archivo_word and os.path.isfile(archivo_word):
            archivo_pdf = os.path.splitext(archivo_word)[0] + ".pdf"
            convert(archivo_word, archivo_pdf)
            estado.set("Conversión completada: " + archivo_pdf)
        else:
            estado.set("Por favor, seleccione un archivo válido.")

    ventana_convertir = tk.Toplevel(root)
    ventana_convertir.title("Conversor de Word a PDF")
    ventana_convertir.attributes("-topmost", True)
    ruta_archivo = tk.StringVar()
    estado = tk.StringVar()

    tk.Label(ventana_convertir, text="Seleccionar archivo de Word:").pack(pady=10)
    tk.Entry(ventana_convertir, textvariable=ruta_archivo, width=50).pack(padx=10)
    

    tk.Button(ventana_convertir, text="Examinar", command=seleccionar_archivo).pack(pady=5)
    tk.Button(ventana_convertir, text="Convertir a PDF", command=convertir_archivo).pack(pady=20)
    tk.Label(ventana_convertir, textvariable=estado).pack(pady=10)

# Crear la ventana principal
root = tk.Tk()
root.title("Menú Principal")
root.geometry("700x600")
root.attributes("-topmost", True)

# Definir colores
color_primario = "#2c3e50"
accent_color = "#3498db"
texto_color = "#ecf0f1"
hover_color = "#2980b9"
root.configure(bg=color_primario)

# Configurar estilos
style = ttk.Style()
style.theme_use('clam')
style.configure('TFrame', background=color_primario)
style.configure('TButton', 
                font=('Helvetica', 12, 'bold'),
                background=accent_color,
                foreground=texto_color,
                padding=10,
                width=25)
style.map('TButton', 
          background=[('active', hover_color)],
          foreground=[('active', texto_color)])

# Crear el marco principal
main_frame = ttk.Frame(root, padding="20 20 20 20")
main_frame.pack(expand=True, fill='both')

# Título
title_label = ttk.Label(main_frame, 
                        text="HERRAMIENTA PARA FOLIAR DOCUMENTOS",
                        font=('Helvetica', 18, 'bold'),
                        foreground=texto_color,
                        background=color_primario)
title_label.pack(pady=(0, 30))

# Cargar y redimensionar la imagen
imagen_original = Image.open("D:\Foliar\src\\btnCrear.png")
imagen_redimensionada = imagen_original.resize((340, 140))  # Ajusta el tamaño según lo necesario
imagen = ImageTk.PhotoImage(imagen_redimensionada)


btn_crear_doc_foliado = tk.Button(
    main_frame, 
    text="CREAR Y FOLIAR DOCUMENTO 📜",
    image=imagen, 
    command=crear_documento_foliado,
    bg=color_primario,
    width=310,height=64,
    border=0,
    activebackground=color_primario
)
btn_crear_doc_foliado.pack(pady=10)

# Cargar y redimensionar la imagen
#imagen_convertir = Image.open("btnSalir.png")
#imagen_c = imagen_convertir.resize((340, 140))  # Ajusta el tamaño según lo necesario
#imagen_cf = ImageTk.PhotoImage(imagen_c)

btn_convertir_docx_pdf = tk.Button(
    main_frame, 
    text="CONVERTIR DE DOCX A PDF", 
    command=cambiar_formatos,
    
    #,image=imagen_cf
)
btn_convertir_docx_pdf.pack(pady=10)

# Cargar y redimensionar la imagen
imagen_salir = Image.open("btnSalir.png")
imagen_r = imagen_salir.resize((340, 140))  # Ajusta el tamaño según lo necesario
imagen_s = ImageTk.PhotoImage(imagen_r)


btn_salir = tk.Button(
    main_frame, 
    text="SALIR", 
    command=root.destroy,
    image=imagen_s,
    bg=color_primario,
    border=0,
    width=310,height=64,
    activebackground=color_primario
)
btn_salir.pack(pady=10)



root.mainloop()

```
